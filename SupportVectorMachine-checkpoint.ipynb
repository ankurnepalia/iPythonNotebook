### Support Vector Machines

There are many __Supervised Learning__ algorithms but one of the most powerful and interesting models is __Support Vector Machines (SVM's)__


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# use seaborn plotting defaults
import seaborn as sns; sns.set()
```

### Motivating SVM

SVMs are a powerful supervised learning algorithm used for __classification__ or for __regression__. SVMs are a __discriminative__ classifier, i.e. they draw a decision boundary between cluster of data.

Let's create a dataset to show SVM model.

```python
from sklearn.datasets.samples_generator import make_blobs
X, y = make_blobs(n_samples=50, centers=2, random_state=0, cluster_std=0.60)
plt.scatter(X[:,0], X[:, 1], c=y, s=50, cmap='spring')
```

A discriminative classifier attempts to draw a line between the two sets of data. There can be several possibilities which perfectly discriminate between the two sets of data, but we need to find a line that signifies the largest margin between these two classes.

```python
xfit = np.linspace(-1, 3.5)
plt.scatter(X[:,0], X[:,1], c=y, s=50, cmap='spring')

for m, b in [(1,0.65), (0.5, 1.6), (-0.2, 2.9)]:
    plt.plot(xfit, m * xfit + b, '-k')

plt.xlim(-1,3.5)
```

These are three very different separaters which perfectly discriminates betweeen these samples. Depending on which one we choose, a new data point will be classified almost entirely differently. 

How can we pick the right seperator ?

### SVM: Maximizing the margin

SVM are one way to address this kind of problems. SVM not only draw the separator lines but also make sure the margins around the separator line are maximized. Let's demonstrate this with another plot.

```python
xfit = np.linspace(-1, 3.5)
plt.scatter(X[:,0], X[:,1], c=y, s=50, cmap='spring')

for m, b, d in [(1, 0.65, 0.33), (0.5, 1.6, 0.55), (-0.2, 2.9, 0.2)]:
    yfit = m * xfit + b
    plt.plot(xfit, yfit, '-k')
    plt.fill_between(xfit, yfit - d, yfit + d, edgecolor='blue', color='grey', alpha=0.4)
    
plt.xlim(-1, 3.5)    
```

Notice here that if we want to maximize this width, the middle fit is clearly the best. This is the intution of __SVM__, which optimize a linear discriminant model in conjunction with a __margin__ representing the perpendicular distance between the datasets.
