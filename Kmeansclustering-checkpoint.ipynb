## K-Means Clustering (Unsupervised Learning)

Current notebook will explore one of the most commonly used unsupervised ML model, K-Means clustering. 


```python
# Let's import standard python libraries
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns; sns.set()
```

### Introducing K-Means

K means is an unsupervised learning algorithm, which is useful in finding clusters in your data based on data attributes along and does not depend on labels(target variables).

#### K-Means Algorithm

1. X = Training set x1, x2,...,xm
2. Randomly pick K training examples and initialize these K cluster centroids mu1, mu2,...,muk to these K training examples.
3. Initialize m to # of training examples. 
4. Repeat {
            [Cluster Assignment Step]
            for i = 1 to m
            C(i) = index (1 to K) of cluster centroid closest to x(i)
            [Move Centroid Step]
            for k = 1 to K
            mu(k) = average (mean) of points assigned to cluster k
          }
          

```python
from sklearn.datasets.samples_generator import make_blobs
X, y = make_blobs(n_samples=300, centers=4, random_state=0, cluster_std=0.60)
plt.scatter(X[:,0], X[:,1], s=50);
```

For this sample dataset it's relatively easy to figure out that there are 4 possible clusters, BUT if you were to perform an exhaustive search for the different segmentations of the data, the search space would be exponential in the number of points. Fortunately, there is a well-known procedure named **Expectation Maximization (EM)**, which can be used to solve K-Means relatively quickly.

```python
from sklearn.cluster import KMeans
K = KMeans(4) # 4 clusters
K.fit(X)
y_kmeans = K.predict(X)
plt.scatter(X[:,0], X[:,1], c=y_kmeans, s=50, cmap='rainbow')
```

### K-Means Algorithm: Expectation Maximization

K-Means is an example of an algorithm which uses an Expectation-Maximization(EM) approach to arrive at the solution. This 2 step approach works as follows:

1. Guess some cluster centroids
2. Repeat until converged:
    a. Assign points to nearest cluster centeroid
    b. Move the new cluster centroid to mean of points belonging to same cluster. 


### K-Means Caveats

The convergence of this algorithm is not guarenteed. In order to get past local manimum problem, scikit-learn by default uses a large number of random initializations and finds the best results.

Also, to get best results, it's advicable to set number of clusters beforehand. 


## K-Means on Digits

Let's apply K-means to digits dataset to find what these clusters represents. 

```python
from sklearn.datasets import load_digits
digits = load_digits()

K = KMeans(n_clusters = 10)
clusters = K.fit(digits.data)
K.cluster_centers_.shape
```

We see 10 clusters in 64 dimensions. Let's visualize each of these clusters to see what they represent.

```python
fig = plt.figure(figsize=(8,3))
for i in range(10):
    ax = fig.add_subplot(2, 5, 1 + i, xticks=[], yticks=[])
    ax.imshow(K.cluster_centers_[i].reshape((8, 8)), cmap=plt.cm.binary)
```

We see that even without the labels, K-Means is able to find clusters whose means are recognizable digits.

Now, let's fix the permuted cluster labels.

```python
from scipy.stats import mode

labels = np.zeros_like(clusters)
for i in range(10):
    mask = (clusters == i)
    labels[mask] = mode(digits.target[mask])[0]
    
```

Let's use PCA visualization and look at the true cluster labels and K-means cluster labels.

```python
from sklearn.decomposition import PCA

X = PCA(2).fit_transform(digits.data)

kwargs = dict(cmap = plt.cm.get_cmap('rainbow', 10),
              edgecolor='none', alpha=0.6)
fig, ax = plt.subplots(1, 2, figsize=(8, 4))
ax[0].scatter(X[:, 0], X[:, 1], c=labels, **kwargs)
ax[0].set_title('learned cluster labels')

ax[1].scatter(X[:, 0], X[:, 1], c=digits.target, **kwargs)
ax[1].set_title('true labels');
```
